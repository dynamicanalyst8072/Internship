Write a python program to display all the header tags from wikipedia.org.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
url_link = "https://en.wikipedia.org/wiki/Main_Page"
request = requests.get(url_link,'lxml')
soup = BeautifulSoup(request.text,'html')
headers = ['h1','h2','h3','h4','h5','h6']
for data in soup.find_all(headers):
    print(data.name + '->' + data.text)
```

Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
and make data frame.

```python
url_link = 'https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc'
request = requests.get(url_link,'lxml')
soup = BeautifulSoup(request.text,'html.parser')
final_data = []
for data in soup.find_all('div',{'class':'lister-item'}):
    movie_data = {}
    movie_data['movie_name'] = data.find('img')['alt']
    movie_data['release_year'] = data.find('span',{'class':'lister-item-year'}).getText()[1:-1]
    movie_data['runtime'] = data.find('span',{'class':'runtime'}).getText()
    movie_data['genre'] = data.find('span',{'class':'genre'}).getText().replace("\n","")
    movie_data['rating'] = data.find('span',{'class':'rating-rating'}).getText()
    final_data.append(movie_data)
data = pd.DataFrame(final_data)
data
```

Write a python program to scrape product name, price and discounts from https://meesho.com/bags-
ladies/pl/p7vbp .

```python
url_link = 'https://meesho.com/bags-%20ladies/pl/p7vbp'
request = requests.get(url_link,'lxml')
soup = BeautifulSoup(request.text,'html.parser')
products = []
for data in soup.find_all('div',{'class':'sc-dkPtRN ProductList__GridCol-sc-8lnc8o-0 FjWWx jMkQHN'}):
    product_data = {}
    product_data['product_name'] = data.find('p',{'class':'Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS'}).getText()
    product_data['price'] = data.find('h5',{'class':'Text__StyledText-sc-oo0kvp-0 hiHdyy'}).getText()[1:]
    product_data['discount in %'] = data.find('span',{'class':'Text__StyledText-sc-oo0kvp-0 lnonyH'}).getText().split('%')[0]
    products.append(product_data)
pd.DataFrame(products)
```

    a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
    b) Top 10 ODI Batsmen along with the records of their team and rating.

Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
a) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.
b) Top 10 women’s ODI Batting players along with the records of their team and rating.
c) Top 10 women’s ODI all-rounder along with the records of their team and rating.

```python
url_link = 'https://www.icc-cricket.com/rankings/womens/team-rankings/odi'
request = requests.get(url_link,'lxml')
soup = BeautifulSoup(request.text,'html.parser')
team_data = {}
teams = []
data = soup.find('tr',{'class':'rankings-block__banner'})
team_data['team_name'] = data.find('span',{'class':'u-hide-phablet'}).text
team_data['team_ranking'] = data.find('td',{'class':'rankings-block__banner--pos'}).text
team_data['matches'] = data.find('td',{'class':'rankings-block__banner--matches'}).text
team_data['points'] = data.find('td',{'class':'rankings-block__banner--points'}).text
team_data['rating'] = data.find('td',{'class':'rankings-block__banner--rating'}).text.strip()
teams.append(team_data)
for data in soup.find_all('tr',{'class':'table-body'}):
    team_data = {}
    team_data['team_ranking'] = data.find('td',{'class':'table-body__cell--position'}).text
    team_data['team_name'] = data.find('span',{'class':'u-hide-phablet'}).text
    total_data = data.find_all('td',{'class':'table-body__cell'})
    team_data['matches'] = total_data[2].text
    team_data['points'] = total_data[3].text
    team_data['rating'] = total_data[4].text
    teams.append(team_data)
pd.DataFrame(teams)
```

Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content
and the code for the video from the link for the youtube video from the post.

```python
url_link = 'https://coreyms.com/'
request = requests.get(url_link,'lxml')
soup = BeautifulSoup(request.text,'html.parser')
final_data = []
for data in soup.find_all('article'):
    article = {}
    article['heading'] = data.find('h2',{'class':'entry-title'}).text
    article['time'] = data.find('time',{'class':'entry-time'}).text
    article['content'] = data.find('div',{'class':'entry-content'}).text.strip()
    final_data.append(article)
pd.DataFrame(final_data)
```

Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
url_link = 'https://www.dineout.co.in/delhi-restaurants/buffet-special'
request = requests.get(url_link,'lxml')
soup = BeautifulSoup(request.text,'html.parser')
restaurants = []
for data in soup.find_all('div',{'class':'restnt-card restaurant'}):
    restaurant = {}
    restaurant['name'] = data.find('a',{'class':'restnt-name'}).text
    restaurant['cuisine'] = data.find('span',{'class':'double-line-ellipsis'}).text.split('|')[1]
    restaurant['location'] = data.find('div',{'class':'restnt-loc'}).text
    restaurant['rating'] = data.find('div',{'class':'restnt-rating rating-4'}).text
    restaurant['src'] = data.find('img').get('data-src')
    restaurants.append(restaurant)
pd.DataFrame(restaurants)
```

Write a python program to scrape first 10 product details which include product name , price , Image URL from
https://www.bewakoof.com/women-tshirts?ga_q=tshirts

```python
url_link = 'https://www.bewakoof.com/women-tshirts?ga_q=tshirts'
request = requests.get(url_link,'lxml')
soup = BeautifulSoup(request.text,'html.parser')
count = 1
products = []
while(count<11):
    product = {}
    box = soup.find('div',{'id':'testProductcard_'+str(count)})
    product['name']= box.find('h3').get('id')
    product['price'] = box.find('span',{'class':'discountedPriceText'}).text
    product['image'] = box.find('img').get('src')
    count+=1
    products.append(product)
    
pd.DataFrame(products)
    
```
